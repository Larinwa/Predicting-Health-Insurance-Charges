```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn #this is where our neural network module is gonna be
import torch.optim as optim #all of our optimizers are gonna be here

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import root_mean_squared_error
```

### 1. Data Preparation

```python
#load the data
df = pd.read_csv('cleaned_insurance.csv')
```

```python
df
```

```python
df.drop(df.columns[0], axis = 1, inplace = True)
```

```python
df
```

```python
y = df['charges']
```

```python
y = y.values
```

```python
y
```

```python
X = df.drop(columns = ['charges'])
```

```python
X
```

```python

#scale the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

#split the data

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 0) #This by default will set the test size to 0.25

```

```python
# convert the data arrays to tensors

X_train_tensor = torch.tensor(data = X_train, dtype= torch.float32)
y_train_tensor = torch.tensor(data = y_train, dtype= torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(data = X_test, dtype= torch.float32)
y_test_tensor = torch.tensor(data = y_test, dtype= torch.float32).view(-1,1)
```

```python
y_train_tensor
```

### 2. Create the neural network graph

```python


#create a class that will inherit from nn.module

class InsuranceModel(nn.Module):

    # define the constructor
    def __init__(self, input_dim)-> None:
        super().__init__()
        self.input_layer = nn.Linear(in_features= input_dim, out_features = 27)
        self.first_hidden = nn.Linear(in_features= 27, out_features =17 )
        self.second_hidden = nn.Linear(in_features= 17, out_features = 10)
        self.third_hidden = nn.Linear(in_features= 10, out_features = 7)
        #self.fourth_hidden = nn.Linear(in_features= 3, out_features = 2)
        self.output_layer = nn.Linear(in_features= 7, out_features = 1)

    # Create the forward pass
    def forward(self, X:torch.tensor):
        result = torch.relu(self.input_layer(X))
        result = torch.relu(self.first_hidden(result))
        result = torch.relu(self.second_hidden(result))
        result = torch.relu(self.third_hidden(result))
        #result = torch.relu(self.fourth_hidden(result))
        result = self.output_layer(result)

        return result
```

```python
X_train.shape
```

### 3. Initialize the model

```python
# init the model

model = InsuranceModel(input_dim = X_train.shape[1]) #this will extract the columns of x_train

# init the criterion
criterion = nn.MSELoss()

#init the optimizer
optimizer = optim.Adam(params = model.parameters(),
                      lr = 0.0001, weight_decay=1e-8)
```

### 4. Begin the training loop and  5. Evaluation

```python
# choose an epoch

EPOCHS = 6000

for epoch in range(EPOCHS):
    model.train() #put the model in train mode
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output, y_train_tensor)#compute the loss
    loss.backward()#ck propagate the loss
    optimizer.step()

    if epoch % 1000 == 0:
        print(f'Loss: {loss} at epoch: {epoch}')

#evaluation
model.eval()
with torch.no_grad():
    preds = model(X_test_tensor) #get predictions
    preds = preds.numpy().flatten()
    rmse = root_mean_squared_error(y_test, preds)

    print(f'test Final rmse: {rmse}' )

#evaluation
model.eval()
with torch.no_grad():
    preds1 = model(X_train_tensor) #get predictions
    preds1 = preds1.numpy().flatten()
    rmse1 = root_mean_squared_error(y_train, preds1)

    print(f'train Final rmse: {rmse1}' )

```
